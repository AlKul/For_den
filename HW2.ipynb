# №1 (Вариант 4)

Задания: построить график функции одной переменной, x $\in$ [-5;5].
Оформить графики

```python
import matplotlib.pyplot as plt
import numpy as np
```

Введем функцию

```python
def func(x):
    return 2*(np.cos(x) + 2) ** (np.sin(x) - 2.1)
```

Строим график

```python
x = np.linspace(-5, 5, 1000)
y = func(x)
plt.figure(figsize = (12,6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.show()
```

# №3

построить поверхности для x$\in$ [–2,2] и y $\in$ [–2,2], изменяя
значения с шагом 0,5:

$z=(0,5x–3)^2+(0,5y–3)^2$

```python
4 / 0.5
```

```python
def z_func(x, y):
    return (0.5*x - 3)**2+(0.5*y - 3)**2
```

```python
from matplotlib import cm
```

```python
x = np.linspace(-2, 2, 9)
X, Y = np.meshgrid(x, x)
Z = z_func(X, Y)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, 
                       linewidth=0)
```
