Задания: построить график функции одной переменной, x $\in$ [-5;5].

Оформить графики

```python
import matplotlib.pyplot as plt
import numpy as np
```

Введем функцию

```python
def func(x):
    y = -0.00388*x**5+0.005*x**4+0.28*x**3-1.03*x**2-4.61*x+4.78
    return y
```

Строим график

```python
x = np.linspace(-5, 5, 1000)
y = func(x)
plt.figure(figsize = (12,6))
plt.plot(x, y, color = 'red')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('График')
plt.show()
```

Задания: построить 2 графика в одном окне, так, чтобы у каждого
графика была своя область построения. Оформить графики

x $\in$ [–2;2], 

шаг 0,2:

y = 2sin(x)cos(x);

z = 3cos2(x)sin(x).

```python
def y_func(x):
    y = 2*np.sin(x)*np.cos(x)
    return y

def z_func(x):
    z = 3*np.cos(x)**2*np.sin(x)
    return z
```

```python
4 / 0.2
```

```python
x = np.linspace(-2, 2, 21)
y = y_func(x)
z = z_func(x)
plt.figure(figsize = (20,6))
plt.subplot(121)
plt.plot(x, y, '-', color = 'green')
plt.title('y(x)')
plt.grid()
plt.subplot(122)
plt.plot(x, z, '--', color = 'b')
plt.title('z(x)')
plt.grid()
plt.show()

```

__№3__



построить поверхности для x$\in$ [–2,2] и y $\in$ [–2,2], изменяя
значения с шагом 0,5:

z=0,5(2x-5)3+0,5(3y–5)2


```python
4 / 0.5
```

```python
def z(x, y):
    z = 0.5*(2*x-5)**3+0.5*(3*y-5)**2
    return z
```

```python
from matplotlib import cm
```

```python
x = np.linspace(-2, 2, 9)
X, Y = np.meshgrid(x, x)
Z = z(X, Y)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, 
                       linewidth=0)
```
