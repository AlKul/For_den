# Задание 1
Построить график функции одной переменной, x $\in$ [-5;5].
Оформить графики

```python
import matplotlib.pyplot as plt
import numpy as np
```

Задаем функцию

```python
def func(x):
    return 0.00142*x**5 + 0.006*x**4 + 0.39*x**3 - 1.09*x**2 - 2.81*x + 6.19
```

Строим график

```python
x = np.linspace(-5, 5, 1000)
y = func(x)
plt.figure(figsize = (12,6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.show()
```

# Задание 2

Построить два графика в одном окне, так, чтобы у каждого
графика была своя область построения. Оформить графики

x $\in$ [–3,5; 3,5], шаг 0,5:

y = $|sin(4x)*cos(x)|$

z = $cos(x) * sin(4x)$

```python
def y_func(x):
    return abs(np.sin(4*x)*np.cos(x))

def z_func(x):
    return np.sin(x) * np.cos(2*x)
```

```python
(3.5 - (-3.5)) / 0.5 + 1
```

```python
x = np.linspace(-3.5, 3.5, 15)
y = y_func(x)
z = z_func(x)
plt.figure(figsize = (12,6))
plt.subplot(121)
plt.plot(x, y, '-', color = 'b')
plt.title('y(x)')
plt.grid()
plt.subplot(122)
plt.plot(x, z, '-', color = 'g')
plt.title('z(x)')
plt.grid()
plt.show()

```
