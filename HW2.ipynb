Задания: построить график функции одной переменной, x $\in$ [-5;5].
Оформить графики

```python
import matplotlib.pyplot as plt
import numpy as np
```

Введем функцию

```python
def func(x):
    return 14*np.sin(2*x)/x
```

Строим график

```python
x = np.linspace(-5, 5, 1000)
y = func(x)
plt.figure(figsize = (12,6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.show()
```

# №2

Задания: построить 2 графика в одном окне, так, чтобы у каждого
графика была своя область построения. Оформить графики

x $\in$ [–2,5;2,5], шаг 0,5:

y = 4*x + (4 + x**2) **0.5

z = $sin(x) * cos(2*x)$

```python
def y_func(x):
    return 4*x + (4 + x**2) **0.5

def z_func(x):
    return np.sin(x) * np.cos(2*x)
```

```python
5 / 0.5
```

```python
x = np.linspace(-2.5, 2.5, 11)
y = y_func(x)
z = z_func(x)
plt.figure(figsize = (12,6))
plt.subplot(121)
plt.plot(x, y, '-')
plt.title('y(x)')
plt.grid()
plt.subplot(122)
plt.plot(x, z, '-', color = 'b')
plt.title('z(x)')
plt.grid()
plt.show()

```

# №3

построить поверхности для x$\in$ [–2,2] и y $\in$ [–2,2], изменяя
значения с шагом 0,5:

$x^2/a^2 + y^2/b^2 – z^2/c^2 = 0$


```python
a = 3

b = 2

c = 1
```

```python
4 / 0.5
```

```python
def z_func_pos(x, y):
    return ((x**2/ a**2 + y**2/b**2) * c**2)**0.5
```

```python
def z_func_neg(x, y):
    return -((x**2/ a**2 + y**2/b**2) * c**2)**0.5
```

```python
from matplotlib import cm
```

```python
x = np.linspace(-2, 2, 9)
X, Y = np.meshgrid(x, x)
Z1 = z_func_pos(X, Y)
Z2 = z_func_neg(X, Y)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z1, 
                       linewidth=0)
surf = ax.plot_surface(X, Y, Z2, 
                       linewidth=0)
```
