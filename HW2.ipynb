Задания: построить график функции одной переменной, x $\in$ [-5;5].

Оформить графики

```python
import matplotlib.pyplot as plt
import numpy as np
```

Введем функцию

```python
def func(x):
    y= -0.00333*x**5+0.005*x**4+0.25*x**3-1.20*x**2-1.39*x+2.47
    return y
```

Строим график

```python
x = np.linspace(-5, 5, 1000)
y = func(x)
plt.figure(figsize = (12,6))
plt.plot(x, y, color = 'blue')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('График функции')
plt.grid()
plt.show()
```

Задания: построить 2 графика в одном окне, так, чтобы у каждого
графика была своя область построения. Оформить графики

x $\in$ [0;3.6], 

шаг 0,3:

y = 3sin(2x)*cos(3x)

z = sin(x)*cos(2x)

```python
def y_func(x):
    y = 3*np.sin(2*x)*np.cos(3*x)
    return y

def z_func(x):
    z = np.sin(x)*np.cos(2*x)
    return z
```

```python
3.6 / 0.3
```

```python
x = np.linspace(-2, 2, 13)
y = y_func(x)
z = z_func(x)
plt.figure(figsize = (20,6))
plt.subplot(121)
plt.plot(x, y, '-', color = 'green')
plt.title('y(x)')
plt.grid()
plt.subplot(122)
plt.plot(x, z, '-', color = 'black')
plt.title('z(x)')
plt.grid()
plt.show()

```

```python
x = np.linspace(-2, 2, 9)
X, Y = np.meshgrid(x, x)
Z = z(X, Y)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, 
                       linewidth=0)
```
