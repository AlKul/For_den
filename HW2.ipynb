# 1
Построить график функции одной переменной, x $\in$ [-5;5].
Оформить графики

```python
import matplotlib.pyplot as plt
import numpy as np
```

Задаем функцию

```python
def func(x):
    return 5*(0.4*np.sin(x)+1)**(np.arctan(x-1)-5.5)
```

Строим график

```python
x = np.linspace(-5, 5, 1000)
y = func(x)
plt.figure(figsize = (12,6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('График Функции')
plt.show()
```

# 2

Построить два графика в одном окне, так, чтобы у каждого
графика была своя область построения. Оформить графики

x $\in$ [0; 3], шаг 0,25:

y = 3*x**3 *np.cos(2*x)

z = 3*np.sin(x)

```python
def y_func(x):
    return 3*x**3 *np.cos(2*x)

def z_func(x):
    return 3*np.sin(x)
```

```python
(3. - 0) / 0.25
```

```python
x = np.linspace(0., 3., 13)
y = y_func(x)
z = z_func(x)
plt.figure(figsize = (15,6))
plt.subplot(121)
plt.plot(x, y, '-', color = 'b')
plt.title('y(x)')
plt.grid()
plt.subplot(122)
plt.plot(x, z, '-', color = 'r')
plt.title('z(x)')
plt.grid()
plt.show()

```

$z = (2*x^2 – 4*y^2)/ 2$
